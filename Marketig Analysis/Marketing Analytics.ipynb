```python
original_Marketing_df.rename(mapper = str.strip,axis = 'columns',inplace = True) #strip all column names of any unneccessary whitespaces
```

```python
original_Marketing_df.drop('ID',axis = 1, inplace = True)# drop ID column
```

```python
original_Marketing_df.drop_duplicates(inplace = True)#Delete all duplicates while keeping the first instance of the duplicate
original_Marketing_df.reset_index(drop = True, inplace = True)#reset the index of the dataframe
```

```python
#visually check for any missing values
plt.subplots(figsize = (20,20))

sbn.heatmap(original_Marketing_df.isnull(),yticklabels = False,cbar = False)

plt.show()
```

```python
#checking the rows with nuLL values in the income column
listofNulls = original_Marketing_df[original_Marketing_df['Income'].isnull()].index.tolist()#creating a list of indeces of rows wit null values in income column
subset_with_nullincome = original_Marketing_df.iloc[listofNulls]#slicing all rows with null in columns into one dataframe 
subset_with_nullincome#these customers seem to have some sort of desposable income because in most instances there are purchases made
```

```python
original_Marketing_df.iloc[listofNulls].describe(include = 'all')
```

```python
#removing dollar sign from income column and convert it to a numerical value
def clean_money(x): 
    a = int(x.split('$')[1].split('.')[0].replace(",", ""))
    return a
original_Marketing_df['Income'] = original_Marketing_df['Income'].dropna().apply(clean_money)
original_Marketing_df.head()
```

```python
#replace every null value in the Income column with the median of the column
original_Marketing_df['Income'] = original_Marketing_df['Income'].fillna(original_Marketing_df['Income'].median())
```

```python
#create an age column from the Year_Birth Column and then drop the respective column
def Ageing(year):
    age = 2022 - year
    return age

original_Marketing_df['Age_in_years'] = original_Marketing_df['Year_Birth'].apply(Ageing)
original_Marketing_df.drop('Year_Birth',axis = 1, inplace = True)
original_Marketing_df.head()
```

```python
#convert the 'Dt_Customer' column into a datetime[64] and then devide it into three different columns of day, month and year

original_Marketing_df['Dt_Customer'] = pd.to_datetime(original_Marketing_df['Dt_Customer'],infer_datetime_format=True)
original_Marketing_df['Dt_Customer_Day'] = pd.DatetimeIndex(original_Marketing_df['Dt_Customer']).day
original_Marketing_df['Dt_Customer_Month'] = pd.DatetimeIndex(original_Marketing_df['Dt_Customer']).month
original_Marketing_df['Dt_Customer_Year'] = pd.DatetimeIndex(original_Marketing_df['Dt_Customer']).year
original_Marketing_df['Dt_Customer_Day'] = original_Marketing_df['Dt_Customer_Day'].astype(str)
original_Marketing_df['Dt_Customer_Month'] = original_Marketing_df['Dt_Customer_Month'].astype(str)
original_Marketing_df['Dt_Customer_Year'] = original_Marketing_df['Dt_Customer_Year'].astype(str)
original_Marketing_df.drop('Dt_Customer',axis = 1,inplace = True)
original_Marketing_df.head()
```

```python
#from 'MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts' and 'MntGoldProds', create a column of the total amount each customer spent at the shop
original_Marketing_df['TotalMntSpent'] = original_Marketing_df['MntWines'] + original_Marketing_df['MntFruits'] + original_Marketing_df['MntMeatProducts'] + original_Marketing_df['MntFishProducts']  + original_Marketing_df['MntSweetProducts'] + original_Marketing_df['MntGoldProds'] 
original_Marketing_df.head()
```

```python
#from 'NumDealsPurchases', 'NumWebPurchases','NumCatalogPurchases' and 'NumStorePurchases' create a Coulmn named Total_Number_Purchases
original_Marketing_df['Total_Number_Purchases'] = original_Marketing_df['NumDealsPurchases'] + original_Marketing_df['NumWebPurchases'] + original_Marketing_df['NumCatalogPurchases'] + original_Marketing_df['NumStorePurchases']
original_Marketing_df.head()
```

```python
#Properly rename the varables in the country column from abreviations to full country names
original_Marketing_df['Country'].replace('SP','Spain',inplace=True)
original_Marketing_df['Country'].replace('CA','Central America',inplace=True)
original_Marketing_df['Country'].replace('US','United States of America',inplace=True)
original_Marketing_df['Country'].replace('AUS','Australia',inplace=True)
original_Marketing_df['Country'].replace('GER','Germany',inplace=True)
original_Marketing_df['Country'].replace('IND','India',inplace=True)
original_Marketing_df['Country'].replace('SA','South America',inplace=True)
original_Marketing_df['Country'].replace('ME','Mexico',inplace=True)
original_Marketing_df['Country'].unique()
```

```python
#create a column which categorises each customer into an age group
original_Marketing_df['Age Segmentation'] = pd.cut(original_Marketing_df['Age_in_years'], bins = [0,17,24,34,44,54,65,100], labels = ['Below 18 years','Between 18 and 24','Between 25 and 34','Between 35 and 44',
                                                                  'Between 45 and 54','Between 55 and 64', 'Above 65 years'])
```

```python
# define a function that will check attributes for outliers
from scipy import stats

def outlierCalculator(_seriesobj):
    result = _seriesobj[(np.abs(stats.zscore(_seriesobj) > 3.5))]# check the absolute value of an observation and the check if it is within 3.5 standard devitins from the mean
    return result.index

original_Marketing_df.loc[outlierCalculator(original_Marketing_df['Income'])]
# the value at original_Marketing_df.loc[517,'Income'] seems to be an incorrect input. One is of the opinion that the 
#the cutomer simply pressed six repeatedly beacuse they could not be bothered to entert ther correct entry
#thus one feels the observations cannot be trusted and should be dropped
```
